##  **Recursive Vs Iterative Approach**


üîÑ Recursion and iteration are two different approaches to solving problems in programming.

üëâ Iteration uses loops to repeat a set of instructions until a condition is met. It is often used when the number of repetitions is known or easily determined.

Here's an example of an iterative function that calculates the factorial of a number:

```python
def iterative_factorial(n):
    result = 1  # Initialize a variable named result with a value of 1
    for i in range(1, n+1):  # Loop through numbers from 1 to n (inclusive)
        result *= i  # Multiply the current value of result by the current number in the loop
        print(result)  # Print the current value of result
    return result  # Return the final value of result after the loop is finished
```

```python
iterative_factorial(4)
```

üîÑ Recursion, on the other hand, breaks down a problem into smaller subproblems and solves them recursively. It is often used when the problem can be expressed in terms of smaller subproblems of the same kind.

Here's an example of a recursive function that calculates the factorial of a number:

```python
def recursive_factorial(n):
    if n == 0:  # Base case: if n equals 0
        return 1  # Return 1 (as 0! equals 1)
    else:  # If n is not equal to 0
        # Recursively call the function with n-1 as the argument and multiply the result by n
        return n * recursive_factorial(n-1)
```

```python
recursive_factorial(4)
```

üëâ In this example, the base case is when n equals 0, at which point the function returns 1. Otherwise, it calls itself with a simpler version of the problem (n-1), until the base case is reached.

üîÑ Recursion can be more elegant and concise than iteration, but it can also be less efficient and more memory-intensive. It's important to choose the right approach depending on the problem at hand.

##**Recursion Examples**

**1. Factorial: Write a recursive function to compute the factorial of a given number. The factorial of n is defined as the product of all positive integers from 1 to n.**

```python
def factorial(n):
    if n == 0:  # Base case: if n equals 0
        return 1  # Return 1 (as 0! equals 1)
    else:  # If n is not equal to 0
        # Recursively call the function with n-1 as the argument and multiply the result by n
        return n * factorial(n-1)

# Example usage
print(factorial(5))  # Output: 120 (as 5! equals 120)
```

This code defines a function factorial that calculates the factorial of a given positive integer n using recursion. If n is 0, the function returns 1 (as 0! = 1). Otherwise, the function multiplies n by the factorial of n-1 (i.e., (n-1)!) to get the factorial of n. The example usage shows how to call the function to calculate the factorial of 5 (which is 120).

**2. Fibonacci sequence: Write a recursive function to compute the nth number in the Fibonacci sequence. The Fibonacci sequence is a sequence of numbers where each number is the sum of the two preceding ones, starting from 0 and 1.**

```python
def fibonacci(n):
    if n <= 1:  # Base case: if n is less than or equal to 1
        return n  # Return n (i.e., 0 if n is 0, 1 if n is 1)
    else:  # If n is greater than 1
        # Recursively call the function with n-1 and n-2 as the arguments and add the results
        return fibonacci(n-1) + fibonacci(n-2)

# Example usage
print(fibonacci(7))  # Output: 13 (as the 7th Fibonacci number is 13)
```

This code defines a function fibonacci that calculates the nth Fibonacci number using recursion. If n is 0 or 1, the function returns n (as the 0th and 1st Fibonacci numbers are 0 and 1, respectively). Otherwise, the function recursively calls itself with n-1 and n-2 as the arguments and adds the results to get the nth Fibonacci number. The example usage shows how to call the function to calculate the 7th Fibonacci number, which is 13.

**Both of these problems can also be solved iteratively, but recursive solutions are often more elegant and easier to read and understand. However, it's important to keep in mind that recursive solutions can be less efficient in terms of space usage, so for large inputs, an iterative solution may be preferable.**

##**Space and Time Complexity of Recursive and Iterative approach**



**Iterative factorial function:**

* Time complexity: O(n), where n is the input number.
* Space complexity: O(1), because the function uses a constant amount of space to store the result and loop counter.

**Recursive factorial function:**

* Time complexity: O(n), where n is the input number.
* Space complexity: O(n), because the function creates n function call frames on the call stack to store the intermediate values of the function parameters and local variables.



As you can see, both functions have the same time complexity, but the recursive function has a higher space complexity due to the overhead of creating function call frames on the call stack. This means that for large values of n, the recursive function may run out of stack space and cause a stack overflow error, while the iterative function will continue to work without any issues.

**In general, iterative functions tend to be more efficient in terms of space usage, while recursive functions tend to be more elegant and easier to read and understand. It's important to choose the right approach depending on the specific requirements and constraints of the problem at hand.**






##**Some Common Pitfall around Recursion.** üíÄüíÄüíÄ

üö´ Recursion is a powerful tool in programming, but it can be easy to fall into some common pitfalls. Here are some of the most common mistakes made when working with recursion, and how to avoid them:

* üï≥Ô∏è **Infinite recursion:** This happens when a function calls itself indefinitely without ever reaching the base case. To avoid infinite recursion, make sure that your recursive function has a clear base case that will eventually be reached.

* üß± **Redundant work:** Recursive functions can sometimes perform redundant calculations or operations. To avoid this, try to use memoization or dynamic programming techniques to store intermediate results and avoid repeating work.

* üìà **Stack overflow:** Recursive functions can use up a lot of memory if the function calls itself too many times. To avoid this, consider using an iterative approach or optimizing your recursive function to reduce the number of function calls.

* üìù **Poorly defined base case:** If the base case is not well-defined or does not handle all possible inputs, the function may produce unexpected results or errors. To avoid this, carefully consider all possible inputs and make sure that the base case covers all of them.

* üîÄ **Infinite recursion on invalid input:** Recursive functions can sometimes run into infinite recursion if they are called with invalid input. To avoid this, add input validation checks at the beginning of your function to handle invalid inputs before the recursion begins.

By keeping these common pitfalls in mind and carefully designing your recursive functions, you can avoid common errors and make the most of this powerful programming technique.

### **Write a program to find the factorial of a number using recursion.**

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Taking input from user
num = int(input("Enter a number: "))

# Checking if the input is valid
if num < 0:
    print("Sorry, factorial does not exist for negative numbers")
elif num == 0:
    print("The factorial of 0 is 1")
else:
    print("The factorial of", num, "is", factorial(num))
```

### **Write a program to find the sum of natural numbers up to n using recursion.**

```python
def sum_natural_numbers(n):
    if n == 1:
        return 1
    else:
        return n + sum_natural_numbers(n-1)

# Taking input from user
num = int(input("Enter a number: "))

# Checking if the input is valid
if num <= 0:
    print("Invalid input. Please enter a positive integer.")
else:
    print("The sum of natural numbers up to", num, "is", sum_natural_numbers(num))
```

### **Write a program to print the Fibonacci series up to n terms using recursion.**

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return (fibonacci(n-1) + fibonacci(n-2))

# Taking input from user
num = int(input("Enter the number of terms: "))

# Checking if the input is valid
if num <= 0:
    print("Invalid input. Please enter a positive integer.")
else:
    print("Fibonacci series:")
    for i in range(num):
        print(fibonacci(i))
```

### **Write a program to find all possible permutations of a given string using recursion.**

```python
def permutations(string, left, right):
    if left == right:
        print("".join(string))
    else:
        for i in range(left, right + 1):
            string[left], string[i] = string[i], string[left]
            permutations(string, left + 1, right)
            string[left], string[i] = string[i], string[left]

# Taking input from user
str = input("Enter a string: ")

# Calling the function
permutations(list(str), 0, len(str)-1)
```

### **Write a program to find the nth term in the Tribonacci sequence using recursion.**

```python
def tribonacci(n):
    if n == 0 or n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return tribonacci(n-1) + tribonacci(n-2) + tribonacci(n-3)

# Taking input from user
num = int(input("Enter a number: "))

# Checking if the input is valid
if num < 0:
    print("Invalid input. Please enter a non-negative integer.")
else:
    print("The", num, "th term in the Tribonacci sequence is", tribonacci(num))
```

### **Write a program to find the sum of digits of a given number using recursion.**

### **Write a program to generate all possible permutations of a given string using recursion.**

```python
def permutations(s):
    if len(s) == 0:
        return []
    elif len(s) == 1:
        return [s]
    else:
        result = []
        for i in range(len(s)):
            # Generating permutations of remaining string
            remaining = s[:i] + s[i+1:]
            perms = permutations(remaining)

            # Adding current character to each permutation
            for perm in perms:
                result.append(s[i] + perm)

        return result

# Taking input from user
s = input("Enter a string: ")

# Generating and printing all permutations
print("All possible permutations of", s, "are:")
print(permutations(s))
```

### **Write a program to find the minimum number of steps required to convert one string into another using recursion.**

```python
def min_steps(s1, s2, m, n):
    # If either of the strings is empty, return the length of the other string
    if m == 0:
        return n
    if n == 0:
        return m

    # If last characters of the strings are same, ignore them and find minimum steps for remaining strings
    if s1[m-1] == s2[n-1]:
        return min_steps(s1, s2, m-1, n-1)

    # If last characters of the strings are different, find minimum steps for all possible operations and return minimum of them
    return 1 + min(min_steps(s1, s2, m, n-1), # Insert
                   min_steps(s1, s2, m-1, n), # Remove
                   min_steps(s1, s2, m-1, n-1) # Replace
                   )

# Taking inputs from user
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

# Finding and printing minimum number of steps required to convert s1 to s2
print("Minimum number of steps required to convert", s1, "to", s2, "is:", min_steps(s1, s2, len(s1), len(s2)))
```

### **Write a program to find the longest increasing subsequence of a given sequence using recursion.**

```python
def longest_increasing_subsequence(seq, i, n, last):
    if i == n:
        return 0
    else:
        # If current element is greater than last element, include it in the subsequence
        if seq[i] > last:
            return max(1 + longest_increasing_subsequence(seq, i+1, n, seq[i]), # Include current element
                       longest_increasing_subsequence(seq, i+1, n, last) # Exclude current element
                       )
        else:
            # If current element is less than or equal to last element, exclude it from the subsequence
            return longest_increasing_subsequence(seq, i+1, n, last)

# Taking input from user
seq = list(map(int, input("Enter a sequence of integers: ").split()))

# Finding and printing longest increasing subsequence
print("Longest increasing subsequence is of length", longest_increasing_subsequence(seq, 0, len(seq), float('-inf')))
```

###**Activity 3:**

Imagine you are a math teacher explaining sequences to your class. You want to introduce a new sequence that is similar to the Fibonacci sequence, but with a twist! You tell your students about the Tribonacci sequence.

üî¢ The Tribonacci sequence is a sequence of numbers where each term is the sum of the three preceding terms. The first three terms in the sequence are 0, 1, and 1, and the sequence continues as follows: 0, 0, 1, 1, 2, 4, 7, 13, 24, 44, 81.......

Now write a program using recursion which would return the nth term in the tribonacci series.

```python
# Your Code Here
def tribonaucci(n):
  if n==0 or n==1 or n==2:
    return 0
  elif n==3:
    return 1
  else:
    return tribonaucci(n-1)+tribonaucci(n-2)+tribonaucci(n-3)


tribonaucci(6)
```

### **Space and Time Complexity**

The time complexity of this recursive function is exponential, specifically **O(3^n)**, since it makes three recursive calls for every call except for the base cases.

As for the space complexity, it is also exponential because the function makes a recursive call for every previous value of n, leading to a large number of function calls and an exponential growth in the call stack size. Specifically, the space complexity is **O(n)** due to the depth of the call stack.

Therefore, this implementation is not suitable for large values of n, as it would require an enormous amount of time and memory.
