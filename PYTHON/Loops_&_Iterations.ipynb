üçï Let's say you have a list of pizza toppings, like ["pepperoni", "mushrooms", "olives", "sausage"]. Your task is to print each toppings in a different line individually. What would be your approach?

```python
#Non Coder:

# Define a list of pizza toppings
toppings = ["pepperoni", "mushrooms", "olives", "sausage"]
print(toppings[0])
print(toppings[1])
print(toppings[2])
print(toppings[3])
```

But with knowledge of loops, things are much easier!!

```python
#Coder

# Define a list of pizza toppings
toppings = ["pepperoni", "mushrooms", "olives", "sausage"]

# Loop through each topping in the list and print it out
for topping in toppings:
    print(topping)
```

üîÅ You can also use a loop to add new toppings to the list. Let's say you want to add "bacon", "peppers", and "pineapple". You can use a for loop and the "append" method to add each topping to the list:

```python
# Add new toppings to the list using a loop and the append method
new_toppings = ["bacon", "peppers", "pineapple"]

for topping in new_toppings:
    toppings.append(topping)

# Print out the updated list of toppings
print(toppings)
```

üçï Now your list of pizza toppings will be ["pepperoni", "mushrooms", "olives", "sausage", "bacon", "peppers", "pineapple"].

##Generating numbers- range()

üìù Overview:
range() is a built-in Python function used to generate a sequence of numbers. It is commonly used for looping a specific number of times in a for loop. The range() function returns an object of the range class, which represents a sequence of numbers.

üìù Syntax:
The basic syntax of the range() function is as follows:

range(start, stop,step)




üìù Example:
Here's an example of how to use the range() function to generate a sequence of numbers and loop through them using a for loop:

```python
# Loop through numbers from 0 to 4
for i in range(5):
    print(i)
```

```python
# Loop through numbers from 2 to 8, incrementing by 2 each time
for i in range(2, 9, 2):
    print(i)
```

##Using Breaks & continue

üëã Hi there! Loops and iterations are powerful tools in Python programming that allow you to repeat a set of instructions until a certain condition is met.

Sometimes you might want to exit a loop early or skip over certain iterations. This is where the break and continue statements come in.

üí• The break statement is used to exit a loop prematurely. When the break statement is encountered inside a loop, the loop is immediately terminated and the program execution continues with the next statement outside of the loop.

üîú Here's an example:

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

This code will print the numbers 0 through 4, and then exit the loop when i is equal to 5.



üí´ The continue statement is used to skip over certain iterations of a loop. When the continue statement is encountered inside a loop, the current iteration is skipped and the loop continues with the next iteration.

üîú Here's an example:


```python
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

This code will print all the odd numbers between 0 and 9. The continue statement skips over the even numbers and moves on to the next iteration of the loop.

ü§ù In summary, the break and continue statements are useful tools for controlling the flow of a loop and can help you write more efficient and effective Python code.

##Nested Loops in Python üéØüêç



Imagine you're organizing a chess tournament üèÜ. Your task is to bring together the players whose name start with 'A' or 'B' in one team, and 'C' or 'D' in another team.

In Python, you can nest 'for' loops to do this. Let's take a closer look at how nested loops work using our chess tournament scenario:

```python
tournament = [['Alice', 'Charlie'], ['Bob', 'Diana']]

team1=[]
team2=[]
for group in tournament:    # First Loop/ Outer Loop.
    for player in group:      #  Inner loop.
        if player[0]=='A' or player[0]=='B':
          team1.append(player)
        else:
          team2.append(player)

print(f'Team 1 is: {team1} \n And Team 2 is {team2}')
```

In this example, we have two 'for' loops. The outer loop iterates through the players list, while the inner loop does the same. By using nested loops, we can generate the team of ['Alice', 'Bob']
& ['Charlie', 'Diana'] using some conditions & indexing.

Nested loops may appear intimidating at first, but with practice, you'll find them to be powerful tools for tackling complex problems. Just remember to keep track of the loop variables and their respective ranges to avoid confusion! üòÑ

```python
# List of numbers
numbers = [
    [2, 5, 11, 20, 8],
    [9, 4, 15, 28, 17],
    [1, 6, 21, 18, 3],
    [10, 13, 25, 33, 30],
    [14, 7, 16, 19, 22]
]

# Initialize counters for odd and even numbers
even_count = 0
odd_count = 0

# Iterate through the main list and its sublists using nested loops
for sublist in numbers:
    for number in sublist:
        # Check if the number is odd or even and increment the counters accordingly
        if number % 2 == 0:
            even_count += 1
        else:
            odd_count += 1

# Print the total count of odd and even numbers
print(f"Total even numbers: {even_count}")
print(f"Total odd numbers: {odd_count}")
```

### Lists



Imagine you are a teacher and you need to calculate the average grade for a class of 25 students. You have a list of grades for each student and you need to loop through the list to calculate the average grade for the class.

Here's an example of how to use a for loop with a list to calculate the average grade for the class:

```python
# Create a list of grades for each student
grades = [90, 85, 95, 80, 75, 85, 90, 85, 70, 80,
          90, 75, 80, 85, 90, 95, 80, 75, 85, 90,
          85, 90, 75, 80, 85]

# Use a for loop to calculate the total grade for the class
total_grade = 0
for grade in grades:
    total_grade += grade

# Calculate the average grade for the class
average_grade = total_grade / len(grades)

# Print the average grade for the class
print(f"The average grade for the class is {average_grade}.")
```

In this example, the for loop iterates through each item in the grades list and adds it to the total_grade variable. The average grade is then calculated by dividing the total grade by the number of students, and the result is printed to the console.

## **Best Practices /Tips**

1. **Plan your loops carefully:** Before writing a loop, take a moment to think about the problem you're trying to solve and how you want to iterate through the data. This can help you avoid errors and make your code more efficient.

2. **Use descriptive variable names:** When using loop variables, choose descriptive names that reflect what they represent. This can help make your code more readable and easier to understand.

3. **Use range() when working with indices:** If you need to access specific elements in a list or other iterable, use the range() function to generate a sequence of indices. This can make your code more concise and easier to understand.

4. **Avoid modifying the iterable inside the loop:** When looping through an iterable, avoid modifying the iterable itself, as this can lead to unexpected behavior. Instead, create a copy of the iterable or use a different approach to avoid modifying the original data.

5. **Use list comprehension when appropriate:** When you need to create a new list based on an existing iterable, consider using list comprehension. This can simplify your code and make it more readable.

6. **Keep your loops simple:** As a general rule, it's best to keep your loops simple and easy to understand. Complex nested loops can be difficult to debug and maintain over time.



## **Shortcomings**

1. **Time complexity:** Loops can be time-consuming, especially when working with large datasets or nested loops. It's important to be mindful of the time complexity of your loops and find ways to optimize them when necessary.

2. **Infinite loops:** It's easy to create an infinite loop by mistake, which can cause your program to crash or hang indefinitely. To avoid this, always double-check your loop conditions and ensure that they will eventually exit.

3. **Memory usage:** Depending on how you use loops, they can consume a lot of memory. For example, creating a large list using a loop can quickly consume a lot of memory. To avoid this, consider using generators or other memory-efficient techniques.

4. **Complexity:** Nested loops can quickly become complex and difficult to read, especially if they are deeply nested. It's important to keep your code organized and well-structured to avoid confusion and errors.

5. Limited access to previous iterations: In some cases, you may need to access the results of previous iterations in your loop. However, this can be difficult to do with standard loops. ## **Additional Reading: Loops & Iterations**

### Tuples

Tuples are also iterable objects in Python, which means you can use loops and iterations to access each element in the tuple

Scenario 1: Looping through tuple elements directly üìö

You want to print out all the grades in your tuple. To do this, you can loop through the tuple elements directly:

```python
grades = (90, 78, 85, 92, 76, 88, 79)

for grade in grades:
    print(grade)
```

Scenario 2: Looping through tuple elements using index üè∑Ô∏è

Now, you want to print out the grades along with their index (position). To do this, you can loop through the tuple using the range() function and len() function to get the indices of the tuple elements:

```python
for index in range(len(grades)):
    print(f"Grade {index + 1}: {grades[index]}")
```

Scenario 3: Looping through tuple elements using the enumerate() function üî¢

Alternatively, you can use the built-in enumerate() function to loop through the tuple elements along with their indices in a more concise way:

(You can find more about enumerate() in "Do you know?" section below)

```python
for index, grade in enumerate(grades):
    print(f"Grade {index + 1}: {grade}")
```

Looping through tuples is a convenient way to access and work with elements in an ordered, immutable collection. By using different looping techniques, you can efficiently iterate through tuple elements based on your needs. üöÄ





### ü§îDo You Know?ü§î

enumerate() is a built-in Python function that allows you to loop through an iterable (like a list, tuple, or string) and keep track of the index (position) of the current item, along with the item itself. It simplifies your code when you need to perform operations that require both the element and its index.

enumerate() takes an iterable as its argument and returns an enumerate object, which is an iterator containing pairs of indices and elements. You can loop through this enumerate object using a for loop.

Here's an example using enumerate() to loop through a list of colors:

```python
colors = ['red', 'green', 'blue', 'yellow']

for index, color in enumerate(colors):
    print(f"Color {index + 1}: {color}")
```
