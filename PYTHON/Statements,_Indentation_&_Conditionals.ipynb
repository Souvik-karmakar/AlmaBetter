## Elif Statement

alif.jpg

👉 Let's talk about the "elif" statement. An "elif" statement allows us to check multiple conditions and execute different blocks of code depending on which condition is true. Here's an example:


````python
age = int(input())

if age >= 18:
    print("You are an adult")
elif age >= 13:
    print("You are a teenager")
else:
    print("You are a child")
````

👉 In this code, we first check if the variable "age" is greater than or equal to 18 using the comparison operator ">=". If that condition is true, we execute the code inside the "if" statement and print "You are an adult".

If the condition is false, we move on to the next condition using the "elif" statement and check if the variable "age" is greater than or equal to 13 using the same comparison operator ">=". If that condition is true, we execute the code inside the "elif" statement and print "You are a teenager".

For example:

````python
age = int(input())

if age >= 18:
    print("You are an adult")
elif age >= 13:
    print("You are a teenager")
else:
    print("You are a child")
````

If both previous conditions are false, we execute the code inside the "else" statement and print "You are a child".

For example:


````python
age = int(input())

if age >= 18:
    print("You are an adult")
elif age >= 13:
    print("You are a teenager")
else:
    print("You are a child")
````

👍 The "elif" statement is useful when we need to check multiple conditions and execute different code depending on which condition is true.

## Nested if Statement

A nested if statement is simply an if statement that is located within the body of another if statement. The nested if statement is executed only if the condition of the outer if statement is true.

Here's an example of a nested if statement:

````python
# asking user for input
age = int(input("Enter your age: "))
height = int(input("Enter your height in cm: "))

# nested if statements to determine eligibility for a rollercoaster ride
if age >= 10:
    print("You are old enough to ride the rollercoaster")
    if height >= 120:
        print("And you meet the height requirement.")
    else:
        print("But you do not meet the height requirement.")
else:
    print("You are not eligible to ride the rollercoaster because you are too young.")
````

In this code, the user is asked to input their age and height. The code then uses nested if statements to determine if the user is eligible to ride a rollercoaster. If the user is at least 10 years old, the code checks if their height is at least 120 cm. If the user meets both requirements, they are deemed eligible to ride the rollercoaster. If the user fails to meet either requirement, they are not eligible to ride.

👍Nested if statements can be useful in situations where you need to perform a more specific check based on a condition that has already been checked.

🚀 By using these different types of conditional statements in Python, we can create more dynamic and flexible programs that can make decisions based on specific conditions.

#### **One-Line Coding in Conditionals**

One-line conditionals are compact expressions that allow you to write simple conditional statements in a single line. They are particularly useful when the conditions are straightforward and the actions are concise. Python provides a shorthand syntax to achieve this, using the ternary operator (also known as the conditional expression).

Ternary Operator Syntax:
```
value_if_true if condition else value_if_false

```
Explanation:

- The condition is evaluated first.
- If the condition is true, the expression returns value_if_true.
- If the condition is false, the expression returns value_if_false.

**Benefits of One-Line Conditionals:**

- One-line conditionals can improve code readability and maintainability, especially when the logic is simple and doesn't require complex actions or multiple nested conditions.
- They can also help reduce the number of lines of code, making your code more concise and easier to understand.



Example:
Let's consider a simple example of using a one-line conditional to find the maximum value between two variables:


````python
# Traditional way
a = 10
b = 15
if a > b:
    max_val = a
else:
    max_val = b

# Using one-line conditional
max_val = a if a > b else b
````



In the one-line conditional, a is assigned to max_val if a is greater than b; otherwise, b is assigned to max_val. The one-line conditional provides a concise and readable way to express the same logic in a single line.


#### **One-Line Coding with Loops (List Comprehensions)**


List comprehensions are a concise way to create lists in Python. They allow you to create new lists by applying expressions to each element in an existing iterable (e.g., list, tuple, or range) while filtering elements that meet specific conditions.

````python
# List comprehension (one-line loop)
# Syntax: [expression for item in iterable if condition]

numbers = [1, 2, 3, 4, 5]
squared_numbers = [num**2 for num in numbers]
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]

even_numbers = [num for num in numbers if num % 2 == 0]
print(even_numbers)  # Output: [2, 4]
````


**Caution and Readability**

While one-line coding can be powerful and concise, it is essential to maintain code readability and avoid excessively complex expressions. Overusing one-liners can make the code less readable and harder to debug. Always strike a balance between brevity and clarity when utilizing one-line coding techniques.

👨‍🏫 As a teacher, it's time to create a grading system for your students in a subject. Here are the rules! 📝

📈 A student's score can range from 0 to 100.

👎 If a student scores below 60, they get an F.

👌 If a student scores between 60 and 69, they get a D.

👍 If a student scores between 70 and 79, they get a C.
