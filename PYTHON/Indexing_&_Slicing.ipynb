## Are Indexing and Slicing operations different for different data types?

You should know that indexing and slicing operations remain mostly the same across different data types in Python. üêç

Whether you're working with a list üìú, tuple üéä, string üî§, or even a NumPy array üßÆ, the basic syntax for indexing and slicing remains consistent.

This consistency makes it easier to work with different data types in Python since you can use the same techniques across multiple data types. üí™

Of course, there may be some differences in behavior depending on the data type, such as whether the data is mutable or immutable. Still, the basic principles of indexing and slicing remain the same. üí°

## **Slicing**

üî™ Slicing is another common operation in Python that allows us to extract a portion of a sequence, such as a string, tuple, or list. Different types of data structures have different slicing methods, but the general syntax is similar for all of them.

**The basic syntax for slicing with a step size is as follows:**

`sequence[start:stop:step]`

Here's what each part means:

* start: The index from which the slicing begins (inclusive).
* stop: The index at which the slicing ends (exclusive).
* step: The number of elements to skip between each included element.

### **Strings**
üç¥ Let's start with a string, which is a sequence of characters. We can use the slice notation [start:end] to extract a portion of the string. Here's an example:

image.png

````python
my_string = "Hello, world!"
print(my_string[0:5])   # Output: "Hello"
print(my_string[7:])   # Output: "world!"
````

üëâ In this example,
* We used slicing to extract the first five characters of the string using my_string[0:5]
* And the characters after the comma using my_string[7:].

## How do we use the **step size**?
Here we have sliced the whole string but with a step size of 2. So the program would output every second element starting from the first

````python
my_string[::2]
````

####Wondering how we can **reverse** using slicing?

Think about it. reversed string a string starting from the last element of the orignal string and going all the way to the first element. **We can reverse a string by setting the step size to -1**

````python
my_string[::-1]   #step size of -1 to reverse the string
````

````python
my_string[::-2]   #step size of -2 would reverse the string skipping one element in between
````

Same concept applies to other data types as well.

### **Tuples**
üì¶ Tuples are another type of sequence in Python that can be sliced using the slice notation. Here's an example:

````python
my_tuple = (1, 2, 3, 4, 5)
print(my_tuple[1:3])   # Output: (2, 3)
print(my_tuple[:2])   # Output: (1, 2)
````

üëâ In this example, we used slicing to extract a portion of the tuple.
* The first slice my_tuple[1:3] extracts the second and third elements of the tuple.
* The second slice my_tuple[:2] extracts the first two elements of the tuple.

### **Lists**
üçΩÔ∏è Lists are another type of sequence in Python that can be sliced using the slice notation. Here's an example:

````python
my_list = [1, 2, 3, 4, 5]
print(my_list[2:])   # Output: [3, 4, 5]
print(my_list[:3])   # Output: [1, 2, 3]
print(my_list[1:4:2])   # Output: [2, 4]
````

üëâ In this example, we used slicing to extract a portion of the list.
* The first slice my_list[2:] extracts all elements after the second element of the list.
* The second slice my_list[:3] extracts the first three elements of the list.
* The third slice my_list[1:4:2] extracts every second element between the second and fifth elements of the list.

### **Dictionary**


üìö Dictionaries are a collection of key-value pairs in Python, and we can slice a dictionary using a list of keys. However, unlike lists or tuples, slicing with dictionaries is not done using the : operator.

Instead, you can use the dict() constructor to create a new dictionary from a slice of an existing dictionary.

We will explore this when we focus on dictionaries later.

##**Indexing & Slicing Dictionaries.**

Indexing in a dictionary is done using the keys. You can access the value associated with a particular key by using the square bracket notation with the key inside the brackets. For example:

````python
my_dict = {'a': 1, 'b': 2, 'c': 3}
value_a = my_dict['a']  # returns 1
value_a
````

Slicing a dictionary is not possible since dictionaries are unordered and do not support slicing. However, you can get a subset of a dictionary by creating a new dictionary with only the desired keys. You can use the dict() constructor and a list comprehension to achieve this. For example:

````python
my_dict = {'a': 1, 'b': 2, 'c': 3}
subset_dict = dict((key, my_dict[key]) for key in ['a', 'c']) #This is an advanced technique which you'll explore later. Don't worry if you don't understand this yet.
# returns {'a': 1, 'c': 3}
subset_dict
````

This creates a new dictionary subset_dict that contains only the key-value pairs for the keys 'a' and 'c'.

We have utilized list comprehension and iterative loops to execute slicing. However, don't worry if you're not yet familiar with these methods as you will learn about them later.

  - Negative indexing is not limited to accessing elements from the end of a sequence; it also allows you to perform reversed slicing effortlessly! üò≤üîô By using negative indices with slicing, you can quickly extract a reversed subset of your data. Behold:

````python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]

# Reverse the list using negative stride
reversed_fruits = fruits[::-1]
print(reversed_fruits)  # Output: ['elderberry', 'date', 'cherry', 'banana', 'apple']
````

**Multi-dimensional Slicing: Navigating in the Matrix üååüî¢**
  - Python's slicing doesn't stop at one-dimensional sequences; you can masterfully slice through multi-dimensional data structures like a seasoned explorer! üß≠üó∫Ô∏è Let's delve into the world of 2D lists (matrices) and see how slicing works:

````python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Extract the second row using simple indexing
row_2 = matrix[1]
print(row_2)  # Output: [4, 5, 6]

# Extract the second column using list comprehension
column_2 = [row[1] for row in matrix]
print(column_2)  # Output: [2, 5, 8]

# Extract a 2x2 submatrix from the main matrix
submatrix = [row[1:3] for row in matrix[1:3]]
print(submatrix)  # Output: [[5, 6], [8, 9]]
````

**Slice Assignment: Altering Data with Precision üîßüõ†Ô∏è**
  - Did you know that you can use slicing to modify portions of your data in place? Python allows slice assignment, enabling you to replace specific elements without creating a new list or string! üîÑüî†

````python
sentence = "Hello, world!"

# Replace "world" with "Python" using slice assignment
sentence_chars = list(sentence)  # Convert the string to a list for slice assignment
sentence_chars[7:12] = "Python"
new_sentence = "".join(sentence_chars)
print(new_sentence)  # Output: "Hello, Python!"
````
