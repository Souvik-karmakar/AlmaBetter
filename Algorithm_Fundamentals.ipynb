## **Bubble Sort**

Bubble_sort_animation.gif

 Bubble Sort is a simple sorting algorithm that works by repeatedly comparing two adjacent elements and swapping them if they are in the wrong order. It is also known as the sinking sort. üåä

üë• It can be visualized as a queue where people arrange themselves by swapping with each other so that they all can stand in ascending order of their heights. üë´

‚è∞ Bubble Sort has a time complexity of O(n^2) in the average and worst cases scenarios and O(n) in the best-case scenario.

üîÑ We compare two adjacent elements and see if their order is wrong, if the order is wrong we swap them. (i.e arr[i] > arr[j] for 1 <= i < j <= s; where s is the size of the array, if array is to be in ascending order, and vice-versa).

Let's consider an example to understand how bubble sort works. Suppose we have an unsorted list of integers: [5, 2, 7, 3, 1, 8, 6, 4].



First Pass:

Starting from the first element of the list, we compare the first two elements (5 and 2). Since 2 is smaller than 5, we swap them to get [2, 5, 7, 3, 1, 8, 6, 4]. We then compare the next two elements (5 and 7), which are already in the correct order, so we move on to the next pair. We continue this process until we reach the end of the list, which gives us [2, 5, 3, 1, 7, 6, 4, 8]. Here, the largest element (8) has "bubbled" to the end of the list.

Second Pass:

We start again from the first element and compare the first two elements (2 and 5), which are already in the correct order. We then compare the next two elements (5 and 3) and swap them to get [2, 3, 5, 1, 7, 6, 4, 8]. We continue this process until we reach the end of the list, which gives us [2, 3, 1, 5, 6, 4, 7, 8]. Here, the second-largest element (7) has "bubbled" to the second-to-last position.

Third Pass:

We start again from the first element and compare the first two elements (2 and 3), which are already in the correct order. We then compare the next two elements (3 and 1) and swap them to get [2, 1, 3, 5, 6, 4, 7, 8]. We continue this process until we reach the end of the list, which gives us [2, 1, 3, 5, 4, 6, 7, 8]. Here, the third-largest element (6) has "bubbled" to the third-to-last position.

Fourth Pass:


We start again from the first element and compare the first two elements (2 and 1) and swap them to get [1, 2, 3, 5, 4, 6, 7, 8]. We continue this process until we reach the end of the list, which gives us [1, 2, 3, 4, 5, 6, 7, 8]. Here, the list is fully sorted in ascending order.

Here is the implemention of above example:

```python
def bubble_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n-i-1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j]


arr = [5, 2, 7, 3, 1, 8, 6, 4]

bubble_sort(arr)
print("Sorted array is:")
for i in range(len(arr)):
    print("%d" %arr[i])
```

This shows that the bubble sort algorithm has successfully sorted the input array.

## **Linear Search**

üîç Linear search is a basic searching algorithm that works by sequentially checking each element in a collection of data until a match is found or the end of the collection is reached. It's also sometimes called a "sequential search" or "brute force search".


cakesearch.gif

Here's how it works:

1. Start at the beginning of the data set.
2. Compare the first element with the target element being searched for.
3. If the target is found, return the index of the element.
4. If the target is not found, move to the next element in the data set and repeat steps 2 and 3 until the target is found or the end of the data set is reached

Here's an example code in Python for linear search:

```python
# Define the linear search function that takes a list of data and a target value
def linear_search(data, target):
    # Loop over each element in the list of data
    for i in range(len(data)):
        # If the current element is equal to the target value, return the index of the element
        if data[i] == target:
            return i
    # If the target value is not found in the list of data, return -1
    return -1

# Get user input for the list of data and the target value
data = list(map(int, input("Enter the list of data, separated by spaces: ").split()))
target = int(input("Enter the target value: "))

# Call the linear search function with the list of data and the target value
index = linear_search(data, target)

# If the target value is found in the list of data, print a message with the index of the element
if index != -1:
    print(f"Target {target} found at index {index}")
# If the target value is not found in the list of data, print a message indicating it was not found
else:
    print(f"Target {target} not found in data")

```

In this example, the linear_search function takes two arguments: data, which is the list of values to search, and target, which is the value we are searching for. The function uses a for loop to iterate through each element of the list. If the element matches the target, the function returns the index of the element. If no match is found, the function returns -1.

Linear search üîç is a simple and straightforward algorithm, but it can be slow ‚è≥ for large data sets.  Its time complexity is O(n), where n is the size of the data set. For this reason, it's typically only used for small data sets or for cases where the data is unsorted and random access to elements is not possible. ü§î  

For larger data sets, more efficient searching algorithms such as binary search or hash tables are often used instead.

### **Selection Sort**

Selection Sort is another simple sorting algorithm that works by repeatedly finding the minimum element from the unsorted part of an array and putting it at the beginning. üéØ

üë• It can be visualized as a group of people standing in a line and selecting the shortest person and moving them to the front of the line. Then, they repeat the process to find the next shortest person and move them to the front until the entire line is sorted. üë´

‚è∞ Selection Sort has a time complexity of O(n^2) in all cases scenarios, which makes it inefficient for large data sets.

üîç The algorithm iterates through the array, and for each iteration, it finds the minimum element from the unsorted part of the array and swaps it with the first unsorted element. This continues until the entire array is sorted.

For example, consider the following array: [5, 3, 4, 1, 2]

The first iteration finds the minimum element from the unsorted part of the array, which is 1. It swaps 1 with the first element, and the array becomes [1, 3, 4, 5, 2].

The second iteration finds the minimum element from the remaining unsorted part of the array, which is 2. It swaps 2 with the second element, and the array becomes [1, 2, 4, 5, 3].

The third iteration finds the minimum element from the remaining unsorted part of the array, which is 3. It swaps 3 with the third element, and the array becomes [1, 2, 3, 5, 4].

The fourth iteration finds the minimum element from the remaining unsorted part of the array, which is 4. It swaps 4 with the fourth element, and the array becomes [1, 2, 3, 4, 5].

The final iteration finds the minimum element from the remaining unsorted part of the array, which is 64. It swaps 64 with the fifth element, but as it is already the maximum element, no swap is made. The array is now sorted: [1, 2, 3, 4, 5].

Here is the code implementation of this example:

```python
# define a function selection_sort that takes an array as an input
def selection_sort(arr):
    # get the length of the array
    n = len(arr)

    # iterate through the array
    for i in range(n-1):
        # set the minimum index to the current index
        min_index = i

        # iterate through the remaining elements to find the minimum element
        for j in range(i+1, n):
            # if the current element is less than the minimum element, update the minimum index
            if arr[j] < arr[min_index]:
                min_index = j

        # swap the minimum element with the first unsorted element
        arr[i], arr[min_index] = arr[min_index], arr[i]

    # return the sorted array
    return arr

# define an array to be sorted
arr = [5, 3, 4, 1, 2]
print("Given array:", arr)

# call the selection_sort function and assign the sorted array to a new variable
sorted_arr = selection_sort(arr)

# print the sorted array
print("Sorted array is:", sorted_arr)

```

Selection sort üéØ has a time complexity of O(n^2), meaning that its performance decreases as the size of the input list grows. However, selection sort has the advantage of being easy to understand and implement üëç, making it a good choice for small lists or as a building block üî® for more complex sorting algorithms.

### **Merge Sort**

Merge_sort_animation2.gif

Merge Sort is a divide-and-conquer algorithm that works by recursively dividing the input  into two halves, sorting the two halves independently, and then merging them together.

üéØ The main idea behind Merge Sort is to divide the array into smaller subarrays, sort those subarrays, and then merge them back together in the correct order.

üë• It can be visualized as a group of people who are sorting themselves into height order by dividing themselves into two equal groups, sorting each group, and then merging the two sorted groups back together.

‚è∞ Merge Sort has a time complexity of O(nlogn) in all cases scenarios, making it an efficient sorting algorithm for large data sets.

üîç The algorithm works by repeatedly dividing the input array in half until each subarray has only one element. Then, it recursively merges adjacent pairs of subarrays by comparing the smallest element of each subarray and placing the smaller element into a new merged subarray. This process continues until there is only one sorted subarray remaining.

For example, consider the following array: [38, 27, 43, 3, 9, 82, 10]

The Merge Sort algorithm works as follows:

1. Divide the input array into two halves: [38, 27, 43, 3] and [9, 82, 10].

2. Recursively divide each subarray in half until each subarray has only one element: [38], [27], [43], [3], [9], [82], and [10].

3. Merge adjacent pairs of subarrays by comparing the smallest element of each subarray and placing the smaller element into a new merged subarray. The two subarrays [38] and [27] are merged into [27, 38].The two subarrays [43] and [3] are merged into [3, 43]. The two subarrays [9] and [82] are merged into [9, 82].The two subarrays [10] and [3, 43] are merged into [3, 10, 43].

4. Recursively merge adjacent pairs of subarrays until there is only one sorted subarray remaining: [27, 38, 3, 43] and [9, 10, 3, 43, 82].
5. Merge the two sorted subarrays into a single sorted array: [3, 9, 10, 27, 38, 43, 82].

Here's the implementation of this example:

```python
def merge_sort(arr):
    # Check if the input array is only one element or empty, if so, return it
    if len(arr) <= 1:
        return arr

    # Divide the array into two halves
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # Recursively call merge_sort on each half
    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    # Merge the two sorted halves
    i = j = 0  # Initialize two counters for the left and right halves
    merged_arr = []  # Create an empty list to store the merged array
    while i < len(left_half) and j < len(right_half):
        # Compare the first element of each half and add the smaller one to the merged array
        if left_half[i] < right_half[j]:
            merged_arr.append(left_half[i])
            i += 1
        else:
            merged_arr.append(right_half[j])
            j += 1

    # Append any remaining elements from the left or right halves to the merged array
    merged_arr += left_half[i:]
    merged_arr += right_half[j:]

    return merged_arr

arr = [64, 25, 12, 22, 11]
sorted_arr = merge_sort(arr)
print("Sorted array is:", sorted_arr)
```

###**Insertion Sort**

Insertion Sort is another simple sorting algorithm that works by sorting an array one element at a time. It iterates through an array and, for each element, finds the appropriate position within the sorted part of the array and inserts it there. üéØ

üîç The algorithm starts by assuming the first element of the array is already sorted. It then iterates through the remaining unsorted part of the array and for each element, it compares it to the sorted part of the array and inserts it in the correct position. This process repeats until the entire array is sorted.




76_sorting.gif

For example, consider the following array: [64, 25, 12, 22, 11]

The first element is assumed to be sorted, so the algorithm starts with the second element (25) and compares it to the first element (64). Since 25 is smaller, it is moved to the left and the array becomes [25, 64, 12, 22, 11].

The algorithm then moves on to the next unsorted element (12) and compares it to the sorted elements in reverse order. It finds that 12 is smaller than 64, so it moves 64 to the third position and 12 to the second position.

The array now looks like this: [25, 12, 64, 22, 11]

The algorithm continues this process for each unsorted element, comparing it to the sorted elements in reverse order and inserting it into its correct position.

At the end of the process, the array is completely sorted in ascending order: [11, 12, 22, 25, 64].

Here is the implementation for this example:

```python
def insertion_sort(arr):
    n = len(arr)

    # Traverse through 1 to n
    for i in range(1, n):
        key = arr[i]  # Select the key value to be inserted at the correct position
        j = i - 1     # Set the previous element index

        # Move elements of arr[0..i-1], that are greater than key, to one position ahead of their current position
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1

        arr[j+1] = key  # Insert the key value in its correct position

    return arr

arr = [64, 25, 12, 22, 11]
print("Given arrayL", arr)
sorted_arr = insertion_sort(arr)
print("Sorted array is:", sorted_arr)

```


‚è∞ Insertion Sort has a time complexity of O(n^2)  but it can perform better than other algorithms like Selection Sort and Bubble Sort for small data sets.

### **Quick Sort**

Quicksort.gif

Quick sort is a commonly used sorting algorithm that works by dividing an array into smaller sub-arrays and sorting them recursively. It uses a divide-and-conquer approach to sort the elements, making it a fast and efficient algorithm for large datasets.

üë• It can be visualized as a group of people standing in a line and selecting a "pivot" person. Then, they divide the remaining people into two groups, those taller than the pivot and those shorter than the pivot. They repeat this process recursively for each group until the entire line is sorted. üë´

The steps involved in quick sort are as follows:

1. Choose a pivot element from the array.

2. Partition the array so that all elements smaller than the pivot are on its left, and all elements larger than the pivot are on its right.

3. Recursively apply the above two steps to the sub-arrays on the left and right of the pivot until the entire array is sorted.


Here's an implementation of quick sort in Python:



```python
def quick_sort(arr):
    # Base case
    if len(arr) <= 1:
        return arr

    # Selecting the first element as pivot
    pivot = arr[0]

    # Initializing empty arrays to store elements
    # smaller than pivot and greater than pivot
    left = []
    right = []

    # Iterating over the elements in the array,
    # excluding the pivot
    for i in range(1, len(arr)):
        # If the element is smaller than pivot,
        # add it to the left array
        if arr[i] < pivot:
            left.append(arr[i])
        # Otherwise, add it to the right array
        else:
            right.append(arr[i])

    # Recursively sorting the left and right arrays,
    # and concatenating the sorted arrays with the pivot
    return quick_sort(left) + [pivot] + quick_sort(right)

# Testing the function
arr = [65, 33, 54, 1, 12]
print("Given array:", arr)
sorted_arr = quick_sort(arr)
print("Sorted array is:", sorted_arr)
```

In this example, the quick_sort function takes an array arr as input and recursively sorts it using the quick sort algorithm. The base case is when the length of the array is less than or equal to 1, in which case the array is returned as is. The first element of the array is chosen as the pivot, and the remaining elements are partitioned into two sub-lists (left and right) based on whether they are less than or greater than the pivot. The function then recursively sorts the left and right sub-lists and concatenates them with the pivot element in between to get the final sorted array.

### **Binary Search**

üîé Binary search is a searching algorithm that works by dividing the search space in half at each step, using a process of elimination to find the target value. üí° It's typically faster than linear search, especially for large data sets.

üîç Here's how it works:

1. Start by finding the middle element of the data set.
2. Compare the middle element with the target element being searched for.
3. If the target is found, return the index of the element.
4. If the target is less than the middle element, repeat steps 1-3 on the left half of the data set.
5. If the target is greater than the middle element, repeat steps 1-3 on the right half of the data set.
6. If the target is not found, return -1.

binary.gif

Here is an example for Binary Search:

```python
# This function implements binary search algorithm to find the index of the target element in a given list of data.
def binary_search(data, target):
    left = 0   # Initialize the left pointer to the first index of the list.
    right = len(data) - 1   # Initialize the right pointer to the last index of the list.
    while left <= right:   # Loop until left pointer becomes greater than right pointer.
        mid = (left + right) // 2   # Calculate the middle index of the current search space.
        if data[mid] == target:   # If the middle element is the target element, return its index.
            return mid
        elif data[mid] < target:   # If the middle element is smaller than the target element, update the left pointer to mid+1.
            left = mid + 1
        else:   # If the middle element is larger than the target element, update the right pointer to mid-1.
            right = mid - 1
    return -1   # Target element was not found in the list, so return -1.

# Get user input for the list of data and the target value
data = list(map(int, input("Enter the list of data, separated by spaces: ").split()))
target = int(input("Enter the target value: "))

# Call the binary search function with the list of data and the target value
index = binary_search(data, target)

# If the target value is found in the list of data, print a message with the index of the element
if index != -1:
    print(f"Target {target} found at index {index}")
# If the target value is not found in the list of data, print a message indicating it was not found
else:
    print(f"Target {target} not found in data")
```

In this example, the binary_search function takes two arguments: data, which is the list of values to search (which must be sorted), and target, which is the value we are searching for. The function uses a while loop to repeatedly divide the search space in half, comparing the middle element with the target value and adjusting the indices accordingly until the target is found or the search space is empty.



üîç Binary search is a very efficient searching algorithm, with a time complexity of O(log n), where n is the size of the data set. üí® This makes it much faster than linear search for large data sets. üòé However, it requires ‚ö† that the data set be sorted in order for the algorithm to work correctly. üîç If the data set is unsorted, binary search will not be effective and other searching algorithms may be more appropriate.

### **Interpolation Search**

Interpolation search is a üïµÔ∏è‚Äç‚ôÄÔ∏è searching algorithm that improves upon binary search by making use of the distribution  of values in the data set. It is used to search for a target value in a sorted array with uniformly distributed values.

The basic idea üí° of the algorithm is to use interpolation to determine an approximate position  of the target value in the array, and then use binary search to find the exact position.

It calculates the probable position of the üéØ  target value in the sorted array by using the formula:

pos = low + ((target - arr[low]) * (high - low)) / (arr[high] - arr[low])

where "low" is the starting index of the array, "high" is the ending index of the array, "target" is the value being searched, and "arr" is the sorted array.

This formula calculates the position of the üéØ target value in the array based on the assumption that the array has uniformly distributed values. If the values are not uniformly distributed, the performance of the algorithm can degrade.

Once the approximate position of the target value is calculated, binary search is used to find the exact position of the target value. If the target value is found, its index is returned. If it is not found, -1 is returned.

Here's an example of how to use interpolation search with a Python list:

```python
# This function implements Interpolation search algorithm to find the index of the target element in a given list of data.
def interpolation_search(data, target):
    # Set the initial search range
    low = 0
    high = len(data) - 1

    # Continue searching while the search range is valid and the target value might be in the range
    while low <= high and target >= data[low] and target <= data[high]:
        # Calculate the estimated position of the target element using interpolation
        pos = low + ((high - low) // (data[high] - data[low])) * (target - data[low])

        # Check if the estimated position contains the target element
        if data[pos] == target:
            # Return the index of the target element
            return pos
        elif data[pos] < target:
            # If the estimated position is less than the target, search the upper half of the range
            low = pos + 1
        else:
            # If the estimated position is greater than the target, search the lower half of the range
            high = pos - 1

    # If the target element is not found, return -1
    return -1

# Get user input for the list of data and the target value
data = list(map(int, input("Enter the list of data, separated by spaces: ").split()))
target = int(input("Enter the target value: "))

# Call the interpolation search function with the list of data and the target value
index = interpolation_search(data, target)

# If the target value is found in the list of data, print a message with the index of the element
if index != -1:
    print(f"Target {target} found at index {index}")
# If the target value is not found in the list of data, print a message indicating it was not found
else:
    print(f"Target {target} not found in data")

```

In this example, we have a sorted list of numbers, and we want to search for the target value using interpolation search. The algorithm estimates the position of the target element based on the values at the ends of the list and the value of the target itself. It then checks the estimated position and updates the search range accordingly until it finds the target or determines that it doesn't exist.

The time complexity of interpolation search is O(log log n) on average, where n is the size of the array. This is faster than binary search, which has a time complexity of O(log n) in the worst case. However, the worst case time complexity of interpolation search can be O(n) if the values in the array are not uniformly distributed.
